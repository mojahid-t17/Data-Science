```python
# import the dataset 
import seaborn as sns
import pandas as pd
df=sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
cat_cols=df.select_dtypes(include=['object','category']).columns
cat_cols
```

```python
# apply one-hot encoding to categorical columnn day
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
day_encoded=ohe.fit_transform(df[['day']])
day_encoded_df=pd.DataFrame(day_encoded.toarray(),columns=ohe.get_feature_names_out(['day']))
day_encoded_df.head()
```

```python
# add the encoded day colomn in the dataset
df=pd.concat([df,day_encoded_df],axis=1)
df.head()
```

```python
# drop the day colomn
df=df.drop(columns=['day'])
df.head()
```

```python
cat_cols=df.select_dtypes(include=['object','category']).columns
cat_cols
```

```python
#apply label encoding to categorical column

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in cat_cols:
    df[col]=le.fit_transform(df[col])   
df.head()
```

```python
df.info()
```

```python
X=df.drop(columns=['total_bill'])
y=df['total_bill']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)  
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
## SVR- support vector regression 
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(X_train,y_train)
```

```python
y_predict=svr.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
r2=r2_score(y_test,y_predict)
print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R2 Score: {r2}')
```

```python
## apply kernel='linear'
svr_linear=SVR(kernel='linear')
svr_linear.fit(X_train,y_train)
y_predict_linear=svr_linear.predict(X_test)
```

```python
mae_linear=mean_absolute_error(y_test,y_predict_linear)
mse_linear=mean_squared_error(y_test,y_predict_linear)
r2_linear=r2_score(y_test,y_predict_linear)
print(f'Linear Kernel - Mean Absolute Error: {mae_linear}')
print(f'Linear Kernel - Mean Squared Error: {mse_linear}')
print(f'Linear Kernel - R2 Score: {r2_linear}')
```

```python
# hyperparameter tuning of SVR model using GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid={
    'kernel':['rbf','linear'],
    'C':[0.1,1,10,100,1000,0.01,0.001],
    'gamma':[1,100,1000,0.1,0.01,0.001],
    
}
```

```python
grid=GridSearchCV(SVR(),param_grid,verbose=3,refit=True)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid=grid.predict(X_test)
```

```python
mae_grid=mean_absolute_error(y_test,y_pred_grid)
mse_grid=mean_squared_error(y_test,y_pred_grid)
r2_grid=r2_score(y_test,y_pred_grid)
print(f'Grid Search - Mean Absolute Error: {mae_grid}')
print(f'Grid Search - Mean Squared Error: {mse_grid}')
print(f'Grid Search - R2 Score: {r2_grid}')
```
