```python
# decision tree classification on irish dataset
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
import pandas as pd
X=pd.DataFrame(iris.data,columns=iris.feature_names)
X
```

```python
y=iris.target
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# train the data using decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
y_pred
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
acc_score=accuracy_score(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cm)
print(acc_score)
print(cr)
```

```python
# visualization of decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plot_tree(dtc,filled=True,feature_names=iris.feature_names,class_names=iris.target_names)
plt.show()
```

```python
#post-proning for the decision Tree classifier

dtc_pruned=DecisionTreeClassifier(max_depth=2)
dtc_pruned.fit(X_train,y_train)
```

```python
y_pred_post=dtc_pruned.predict(X_test)
```

```python
y_pred_post
```

```python
# confusion matrix for post-pruned decision tree
cm_post=confusion_matrix(y_test,y_pred_post)
acc_score_post=accuracy_score(y_test,y_pred_post)
cr_post=classification_report(y_test,y_pred_post)
print(cm_post)
print(acc_score_post)
print(cr_post)
```
