## Probability Duration Fucntion
probability distribution function describe how the probabilities are distributed over the values of the random variable. The probability distribution function can be used to calculate the probability that a random variable falls within a particular range of values, as opposed to taking on any one value.

### To describe how these variables behave or how their probabilities are distributed, we use three important functions:
1. Probability Mass Function (PMF) - for discrete random variables
2. Probability Density Function (PDF) - for continuous random variables
3. Cumulative Distribution Function (CDF) - for both discrete and continuous random variables
 ### Probability Mass Function (PMF)  
The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. In other words, it provides the probabilities of all possible outcomes of a discrete random variable.
for example, consider a fair six-sided die. The possible outcomes when rolling the die are 1, 2, 3, 4, 5, and 6. The PMF for this random variable can be defined as follows:
- P(X=1) = 1/6
- P(X=2) = 1/6
- P(X=3) = 1/6
- P(X=4) = 1/6
- P(X=5) = 1/6
- P(X=6) = 1/6  
The PMF satisfies the following properties:
1. Non-negativity: P(X=x) >= 0 for all x
2. Normalization: The sum of the probabilities for all possible outcomes equals 1, i.e., Σ P(X=x) = 1
3. Discreteness: The PMF is defined only for discrete random variables.  


### Probability Density Function (PDF)
Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a specific value. Unlike the PMF, which gives the probability of a discrete random variable being exactly equal to a certain value, the PDF provides the relative likelihood of the random variable falling within a particular range of values. The PDF is used for continuous random variables, which can take on any value within a given range.  
The PDF has the following properties:
1. Non-negativity: f(x) >= 0 for all x
2. Normalization: The total area under the PDF curve equals 1, i.e., ∫ f(x) dx = 1 over the entire range of x
3. Probability Calculation: The probability that the random variable falls within a specific interval [a, b] is given by the integral of the PDF over that interval: P(a ≤ X ≤ b) = ∫[ a to b ] f(x) dx 
4. Continuity: The PDF is defined for continuous random variables and can take on any value within a given range.


### Cumulative Distribution Function (CDF)
The Cumulative Distribution Function (CDF) is a function that describes the probability that a random variable takes on a value less than or equal to a specific value. The CDF is applicable to both discrete and continuous random variables.  
The CDF has the following properties:
1. Non-decreasing: F(x) is a non-decreasing function, meaning that as x increases, F(x) does not decrease.
2. Limits: The CDF approaches 0 as x approaches negative infinity and approaches 1 as x approaches positive infinity, i.e., lim (x→-∞) F(x) = 0 and lim (x→∞) F(x) = 1.
3. Right-continuity: The CDF is right-continuous, meaning that for any point x0, the limit of F(x) as x approaches x0 from the right is equal to F(x0).
4. Relationship with PMF and PDF: For discrete random variables, the CDF can be obtained by summing the PMF up to a certain value: F(x) = Σ P(X ≤ x). For continuous random variables, the CDF can be obtained by integrating the PDF up to a certain value: F(x) = ∫[ -∞ to x ] f(t) dt.
The cumulative distribution curve provides a complete description of the distribution of a random variable and is useful for calculating probabilities and quantiles.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for Normal Distribution
mu = 0     # mean
sigma = 1   # standard deviation

# Generate x values (range of possible values)
x = np.linspace(-4, 4, 1000)

# Calculate PDF and CDF
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Create plots
plt.figure(figsize=(8, 3))

# Plot PDF
plt.plot(x, pdf, label='PDF (Probability Density Function)', color='blue', linewidth=2)

# Plot CDF
plt.plot(x, cdf, label='CDF (Cumulative Distribution Function)', color='red', linewidth=2)

# Add labels, title, legend, and grid
plt.title('Normal Distribution: PDF and CDF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
