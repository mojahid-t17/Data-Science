```python
# Create a comprehensive grade management system with error handling

def grade_manager(source_path,destination_path):
    try:
        with open(source_path,'r',encoding='utf-8') as file:
            print(f"Reading from {source_path}.......")
            content=[lines.strip() for lines in file.readlines()]
            # print(content)
            
            lines=[lines for lines in content if not lines.startswith('#')]
            # print(lines)
            total_lines=len(lines)
            
            student_grades={}
            invalid_lines=[]
            for line in lines:
                if "," not in line:
                    Invalid_message=f"line {content.index(line)+1}:invalid format (expected \"Name,Score\")"
                    invalid_lines.append(Invalid_message)
                    print(f"Writing {Invalid_message}")
                    
                    continue
                line=line.split(',')
                # print(line)
                current_line=",".join(line)  
                # print("current line is:",current_line)
                name=line[0]
                # print(name)
                if(name==""):
                    Invalid_message=f"line {content.index(current_line)+1}: Name is Empty"
                    invalid_lines.append(Invalid_message)
                    print(f"writing {Invalid_message}")
                    continue
                try:
                    score=float(line[1])
                except ValueError:
                    Invalid_message=f"line {content.index(current_line)+1}:(score is not a number) "
                    invalid_lines.append(Invalid_message)
                    print(f"writing {Invalid_message}")
                    continue
                else:
                    if(score<0 or score>100):
                        Invalid_message=f"line {content.index(current_line)+1}: Score {score} is outside 0-100"
                        invalid_lines.append(Invalid_message)
                        print(f"writing {Invalid_message}")
                        continue
                # student_grades[name]=score
                # if same name is duplicated
                student_grades[name.strip().lower()]=score
            
            # print(student_grades)
            # print(invalid_lines)
            print(f"\nloaded {len(lines)} line -> valid {len(lines)-len(invalid_lines)}, invalid {len(invalid_lines)}")
            
    
            print("\nStatus: ")
            average=0
            highest_scorer=max(student_grades, key=student_grades.get)
            lowest_scorer=min(student_grades, key=student_grades.get)
            lowest=0
            highest=0
            for key,value in student_grades.items():
                # print(key,value)
                lowest=student_grades[lowest_scorer]
                highest=student_grades[highest_scorer]
                average=average+value /len(student_grades)
            
            print(f"-Average Score: {average:.2f}")
            print(f"-Highest:{highest} ({highest_scorer})")
            print(f"-Lowest: {lowest} ({lowest_scorer})")
            with open(destination_path,'w',encoding="utf-8") as file:
                from datetime import datetime
                current_datetime = datetime.now()
                try:
                    file.write("student grade report\n")
                    file.write("="*20 )
                    file.write(f"\nwrite from file: {source_path}\n")
                    # write the current date and time
                    file.write(f"report generated: {current_datetime.strftime("%Y-%m-%d %H:%M:%S")}\n\n")
                    file.write("Records:\n")
                    file.write("-"*20)
                    file.write(f"\nTotal lines: {total_lines}\n")
                    file.write(f"Valid students: {len(student_grades)}\n")
                    file.write(f"Invalid lines: {len(lines)-len(student_grades)}\n")
                    for line in invalid_lines:
                        file.write(f"-{line}\n")
                    file.write("\nStatistics\n")
                    file.write(f"Average Score: {average:.2f}\n")
                    file.write(f"Highest: {highest}({highest_scorer})\n")
                    file.write(f"Lowest: {lowest}({lowest_scorer})\n")
                    file.write(f"\n\nStudents (sorted by score desc)\n")
                
                
                     # Sort report by score descending
                    sorted_grades = sorted(student_grades.items(), key=lambda x: x[1], reverse=True)
                    index = 1
                    for name, score in sorted_grades:  
                       
                        file.write(f"{index}) {name} : {score:.2f}\n",)
                        index += 1

                    
                    
                except Exception as e:
                    print(e)
                else:
                    print(f"Report written to : {destination_path}")  
                    
                    
                    
            
            
            
    except FileNotFoundError:
        print(f"file {source_path} not found")

grade_manager("grade.txt","report.txt")
```
